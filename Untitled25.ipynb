```python
fake_news_dataset = pd.read_csv('/content/fake.csv')
true_news_dataset = pd.read_csv('/content/true.csv')

fake_news_dataset['label']=1
true_news_dataset['label']=0
```

```python
#combines both datasets into one.
news_dataset = pd.concat([true_news_dataset,fake_news_dataset],ignore_index=True)
news_dataset
```

```python
news_dataset = news_dataset.sample(frac=1).reset_index(drop=True)
news_dataset.head()
```

```python
news_dataset ['content']=news_dataset.title +" " + news_dataset.text
```

```python
news_dataset = news_dataset[['content','label']]
news_dataset.head()
```

```python
#prepare tools for text cleaning.
port_stem = PorterStemmer()
```

```python
port_stem = PorterStemmer()
stop_words = set(stopwords.words('english'))

def stemming(content):
  #remove non-alphabet characters and lowercase
  content = re.sub('[^a-zA-Z]',' ',content).lower().split()

  #Remove stopwards and stem
  content = [port_stem.stem(word) for word in content if not word in stop_words]
  #join the stemmed words into a string
  return ' '.join(content)
```

```python
from tqdm import tqdm##tqdm="taqaddum"(Arabic word meaning "progress")
tqdm.pandas()

news_dataset['content'] = news_dataset['content'].progress_apply(stemming)
#cleans entire dataset.
```
